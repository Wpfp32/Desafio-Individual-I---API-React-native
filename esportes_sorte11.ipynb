```python
import random

def gerar_rodada_pernambucana():
    clubes = {
        "Sport": {"vitorias": 10, "empates": 2, "derrotas": 1, "jogadores": ["Gustavo Coutinho", "Lucas Lima", "Chrystian Barletta"]},
        "Náutico": {"vitorias": 7, "empates": 3, "derrotas": 3, "jogadores": ["Paulo Sérgio", "Patrick Allan", "Thalissinho"]},
        "Santa Cruz": {"vitorias": 5, "empates": 4, "derrotas": 4, "jogadores": ["Pedro Bortoluzo", "Thiaguinho", "Caio Mello"]}
    }

    status_time = ["Em busca da liderança", "G-4 consolidado", "Pressão da torcida", "Em ascensão"]

    print("🔔 ATUALIZAÇÃO DO CAMPEONATO PERNAMBUCANO 🔔\n")

    for time, info in clubes.items():
        # Dados Gerais do Time (Mudam a cada rodada simulada)
        posicao = random.randint(1, 10)
        clima = random.choice(status_time)

        print(f"🏟️ CLUBE: {time.upper()}")
        print(f"📊 Classificação: {posicao}º lugar | Status: {clima}")
        print(f"📈 Campanha: {info['vitorias']}V - {info['empates']}E - {info['derrotas']}D")
        print("-" * 30)

        # Sorteia 1 jogador do elenco para destacar no Card
        p_nome = random.choice(info['jogadores'])
        gols = random.randint(0, 10)
        precisao = f"{random.randint(70, 98)}%"
        odd = round(random.uniform(1.50, 6.00), 2)

        print(f"🌟 DESTAQUE PARA O CARD: {p_nome}")
        print(f"⚽ Gols no Estadual: {gols} | Precisão de Passe: {precisao}")
        print(f"💎 Odd na Esportes da Sorte: {odd}")
        print("=" * 40 + "\n")

# Executa a simulação
gerar_rodada_pernambucana()
```

```python
import random
import json

def gerar_api_mock_pernambucano():
    clubes = {
        "Sport": {"vitorias": 10, "empates": 2, "derrotas": 1, "jogadores": ["Gustavo Coutinho", "Lucas Lima", "Chrystian Barletta"]},
        "Náutico": {"vitorias": 7, "empates": 3, "derrotas": 3, "jogadores": ["Paulo Sérgio", "Patrick Allan", "Thalissinho"]},
        "Santa Cruz": {"vitorias": 5, "empates": 4, "derrotas": 4, "jogadores": ["Pedro Bortoluzo", "Thiaguinho", "Caio Mello"]}
    }

    status_opcoes = ["Em busca da liderança", "G-4 consolidado", "Pressão da torcida", "Em ascensão"]
    resultado_final = []

    for time, info in clubes.items():
        # Dados do Time
        dados_time = {
            "clube": time,
            "classificacao": random.randint(1, 10),
            "status_clima": random.choice(status_opcoes),
            "campanha": f"{info['vitorias']}V - {info['empates']}E - {info['derrotas']}D",
            "cards_jogadores": []
        }

        # Dados de cada Jogador do Time
        for p_nome in info['jogadores']:
            dados_time["cards_jogadores"].append({
                "nome": p_nome,
                "gols_estaduais": random.randint(0, 10),
                "precisao_passe": f"{random.randint(70, 98)}%",
                "odd_esportes_da_sorte": round(random.uniform(1.50, 6.00), 2),
                "posicao": "Destaque"
            })

        resultado_final.append(dados_time)

    # Transforma em formato JSON (o que o Front-end entende)
    json_final = json.dumps(resultado_final, indent=4, ensure_ascii=False)
    return json_final

# Executa e mostra o que o Front-end vai receber
meu_json = gerar_api_mock_pernambucano()
print(meu_json)
```
